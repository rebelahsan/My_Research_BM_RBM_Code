```python
import numpy as np
# Numeric labels for models
model_labels = {'BM Train': 0, 'BM Test': 1, 'RBM Train': 2, 'RBM Test': 3}

# Prepare data for 3D surface plot
epochs_grid, models_grid = np.meshgrid(epochs, list(model_labels.values()))
elbo_values = np.array([
    bm_train_elbo,
    bm_test_elbo,
    rbm_train_elbo,
    rbm_test_elbo
])

# Plotting the 3D surface plot
fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(111, projection='3d')

# Plotting the surface
surf = ax.plot_surface(epochs_grid, models_grid, elbo_values, cmap='viridis', edgecolor='none')

# Adding labels and title
ax.set_xlabel('Epochs')
ax.set_ylabel('Model')
ax.set_zlabel('ELBO')
ax.set_title('3D Surface Plot of ELBO for BM and RBM')

# Customizing the y-axis to show model names
ax.set_yticks(list(model_labels.values()))
ax.set_yticklabels(list(model_labels.keys()))

# Adding a color bar
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# Calculate the rate of change for ELBO values
def calculate_rate_of_change(elbo_values):
    return [(elbo_values[i] - elbo_values[i-1]) / (epochs[i] - epochs[i-1]) for i in range(1, len(elbo_values))]

bm_train_roc = calculate_rate_of_change(bm_train_elbo)
bm_test_roc = calculate_rate_of_change(bm_test_elbo)
rbm_train_roc = calculate_rate_of_change(rbm_train_elbo)
rbm_test_roc = calculate_rate_of_change(rbm_test_elbo)

# Plotting ELBO values
plt.figure(figsize=(14, 6))

# ELBO values
plt.subplot(1, 2, 1)
plt.plot(epochs, bm_train_elbo, marker='o', linestyle='-', color='blue', label='BM Training ELBO')
plt.plot(epochs, bm_test_elbo, marker='o', linestyle='--', color='blue', label='BM Test ELBO')
plt.plot(epochs, rbm_train_elbo, marker='^', linestyle='-', color='red', label='RBM Training ELBO')
plt.plot(epochs, rbm_test_elbo, marker='^', linestyle='--', color='red', label='RBM Test ELBO')
plt.title('ELBO Values over Epochs')
plt.xlabel('Epochs')
plt.ylabel('ELBO')
plt.legend()
plt.grid(True)

# Rate of Change
roc_epochs = epochs[1:]
plt.subplot(1, 2, 2)
plt.plot(roc_epochs, bm_train_roc, marker='o', linestyle='-', color='blue', label='BM Training ELBO ROC')
plt.plot(roc_epochs, bm_test_roc, marker='o', linestyle='--', color='blue', label='BM Test ELBO ROC')
plt.plot(roc_epochs, rbm_train_roc, marker='^', linestyle='-', color='red', label='RBM Training ELBO ROC')
plt.plot(roc_epochs, rbm_test_roc, marker='^', linestyle='--', color='red', label='RBM Test ELBO ROC')
plt.title('Rate of Change of ELBO Values')
plt.xlabel('Epochs')
plt.ylabel('Rate of Change of ELBO')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Creating DataFrame for summary
roc_data = {
    'Epochs': roc_epochs,
    'BM Training ROC': bm_train_roc,
    'BM Test ROC': bm_test_roc,
    'RBM Training ROC': rbm_train_roc,
    'RBM Test ROC': rbm_test_roc
}

roc_df = pd.DataFrame(roc_data)
roc_df
```

```python
import matplotlib.pyplot as plt



# Creating DataFrame for summary
roc_epochs = epochs[1:]
roc_data = {
    'Epochs': roc_epochs,
    'BM Training ROC': bm_train_roc,
    'BM Test ROC': bm_test_roc,
    'RBM Training ROC': rbm_train_roc,
    'RBM Test ROC': rbm_test_roc
}

roc_df = pd.DataFrame(roc_data)

# Plotting ROC graph as bars
plt.figure(figsize=(10, 6))
bar_width = 0.2
index = np.arange(len(roc_epochs))

plt.bar(index, bm_train_roc, bar_width, label='BM Training ROC', color='blue')
plt.bar(index + bar_width, bm_test_roc, bar_width, label='BM Test ROC', color='blue', alpha=0.5)
plt.bar(index + 2 * bar_width, rbm_train_roc, bar_width, label='RBM Training ROC', color='red')
plt.bar(index + 3 * bar_width, rbm_test_roc, bar_width, label='RBM Test ROC', color='red', alpha=0.5)

plt.xlabel('Epochs')
plt.ylabel('Rate of Change of ELBO')
plt.title('Rate of Change (ROC) of ELBO Values over Epochs')
plt.xticks(index + 1.5 * bar_width, roc_epochs)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

roc_df
```

```python
# Re-plotting the ROC values with smooth curves for each line
plt.figure(figsize=(10, 6))

# Creating smooth curves
from scipy.interpolate import make_interp_spline

epochs_new = np.linspace(min(roc_epochs), max(roc_epochs), 300)
bm_train_roc_smooth = make_interp_spline(roc_epochs, bm_train_roc)(epochs_new)
bm_test_roc_smooth = make_interp_spline(roc_epochs, bm_test_roc)(epochs_new)
rbm_train_roc_smooth = make_interp_spline(roc_epochs, rbm_train_roc)(epochs_new)
rbm_test_roc_smooth = make_interp_spline(roc_epochs, rbm_test_roc)(epochs_new)

plt.plot(epochs_new, bm_train_roc_smooth, linestyle='-', color='blue', label='BM Training ROC')
plt.plot(epochs_new, bm_test_roc_smooth, linestyle='--', color='green', label='BM Test ROC')
plt.plot(epochs_new, rbm_train_roc_smooth, linestyle='-', color='red', label='RBM Training ROC')
plt.plot(epochs_new, rbm_test_roc_smooth, linestyle='--', color='purple', label='RBM Test ROC')

plt.title('Rate of Change (ROC) of ELBO Values over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Rate of Change of ELBO')
plt.legend()
plt.grid(True)


#roc_plot_curve_path
```
